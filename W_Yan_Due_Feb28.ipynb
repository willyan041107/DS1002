```python

from english_words import get_english_words_set
import numpy as np
import seaborn as sns
words = get_english_words_set(['web2'], lower=True)
```

```python
# Determine the data type of words
type_of_words = type(words)
print(f"The data type of 'words' is: {type_of_words}")
```

```python
# Convert to list and find the length
wordlist = list(words)
length_of_wordlist = len(wordlist)
print(f"Length of wordlist: {length_of_wordlist}")

# Print out the first ten words
print("First ten words in wordlist:", wordlist[:10])
```

```python
# Create alpha_list and print the first ten words
alpha_list = sorted(wordlist)
print("First ten words in alpha_list:", alpha_list[:10])
```

```python
# Count words starting with 'a'
a_start_count = sum(1 for word in wordlist if word.startswith('a'))
print(f"Number of words starting with 'a': {a_start_count}")
```

```python
# Count words with 'q' and fraction without 'e'
q_count = sum(1 for word in wordlist if 'q' in word)
no_e_fraction = sum(1 for word in wordlist if 'e' not in word) / length_of_wordlist
print(f"Number of words with 'q': {q_count}")
print(f"Fraction of words without 'e': {no_e_fraction}")
```

```python
import numpy as np

# Convert lists to numpy arrays
wordlist_np = np.array(wordlist)
alpha_list_np = np.array(alpha_list)

# Generating an alphabet list
alphabet = [chr(i) for i in range(ord('a'), ord('z')+1)]
print("Alphabet:", alphabet)
```

```python
import matplotlib.pyplot as plt

# Count the number of words starting with each letter
letter_counts = {letter: sum(1 for word in wordlist if word.startswith(letter)) for letter in alphabet}

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(letter_counts.keys(), letter_counts.values())
plt.title('Number of Words Starting with Each Letter')
plt.xlabel('Letter')
plt.ylabel('Number of Words')
plt.show()
```

```python
# Find the six longest words
six_longest_words = sorted(wordlist, key=len, reverse=True)[:6]
print("Six longest words:", six_longest_words)
```

```python
# Histogram of word lengths
word_lengths = [len(word) for word in wordlist]
plt.figure(figsize=(10, 6))
plt.hist(word_lengths, bins=range(1, max(word_lengths)+2), edgecolor='black')
plt.title('Histogram of Word Lengths')
plt.xlabel('Length of Word')
plt.ylabel('Number of Words')
plt.show()
```

```python
# Count words containing both 'v' and 'a', and both 'w' and 'l'
va_count = sum(1 for word in wordlist if 'v' in word and 'a' in word)
wl_count = sum(1 for word in wordlist if 'w' in word and 'l' in word)

# Measurement of execution time is typically done using the time module or magic commands in Jupyter notebooks
print(f"Words containing both 'v' and 'a': {va_count}")
print(f"Words containing both 'w' and 'l': {wl_count}")
```

```python
# Enumerate all the pairs without duplictates or a pair of the same character
pairs = []
for i in range(97, 122+1):
  a = chr(i)
  for j in range(i+1, 122+1):
    b = chr(j)
    pair=[0,0]
    pair[0] = a
    pair[1] = b
    if a != b:
      pairs.append(pair)
    i += 1
```

```python
# Count of every pair
counts = []
for pair in pairs:
  count = sum(1 for word in wordlist if pair[0] in word and pair[1] in word)
  counts.append(count)
```

```python
# Most frequent pair
max_count = max(counts)
max_index = counts.index(max_count)
max_pair = pairs[max_index]
print("The pair of letters that appear the most frequently is: ", max_pair, ",", max_count, "contain both those letters")
```

```python
# Least frequent pair
min_count = min(counts)
min_index = counts.index(min_count)
min_pair = pairs[min_index]
print("The pair of letters that appear the least frequently is: ", min_pair, ",", min_count, "contain both those letters")
```

```python
# Enumerate all the pairs without removing duplicates so that we can get a full heatmap
pairs = []
for i in range(97, 122+1):
  a = chr(i)
  for j in range(97, 122+1):
    b = chr(j)
    pair=[0,0]
    pair[0] = a
    pair[1] = b
    pairs.append(pair)
```

```python
# Count of every pair
counts = []
for pair in pairs:
  count = sum(1 for word in wordlist if pair[0] in word and pair[1] in word)
  counts.append(count)
# Split the list by every character
heat = np.array_split(counts, 26)
```

```python
# Draw the heatmap
sns.heatmap(heat)
```
